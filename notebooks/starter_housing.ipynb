```
for column in housing.columns:
    if column not in ['id', 'date', 'price']:
        # Convert to category for better boxplot visualization if the column has few unique values
        if housing[column].nunique() < 50 and housing[column].dtype != 'object':
            housing[column] = housing[column].astype('category')

        print(f"Generating boxplot for '{column}' vs 'price'...")
        p = (
            ggplot(housing, aes(x=column, y='price')) +
            geom_boxplot(fill="steelblue", color="black") +
            labs(title=f'Price Distribution by {column}', x=column, y='Price') +
            theme(axis_text_x=element_text(rotation=45, hjust=1))
        )
        print(p)

# Special handling for 'date' if needed, though often not ideal for direct boxplot x-axis
# If you wish to analyze date, you might extract year, month, or day of week first.
```

```
# Identify categorical columns to one-hot encode
# Excluding 'id' and 'date' as they require different handling
categorical_cols = ['bedrooms', 'bathrooms', 'floors', 'waterfront', 'view', 'condition', 'grade', 'zipcode']

# Ensure these columns are of 'category' dtype before one-hot encoding, if not already
for col in categorical_cols:
    if col in housing.columns:
        housing[col] = housing[col].astype('category')

# Perform one-hot encoding
housing_encoded = pd.get_dummies(housing, columns=categorical_cols, drop_first=True)

# Display the first few rows and the shape of the new DataFrame to show the effect of encoding
print("Shape of DataFrame after one-hot encoding:", housing_encoded.shape)
display(housing_encoded.head())
```

```
top_feats = ['price', 'sqft_living', 'sqft_lot', 'bedrooms', 'bathrooms', 'grade', 'sqft_living15']
sns.pairplot(housing[top_feats], diag_kind='kde')
plt.show()
```

```
plt.figure(figsize=(8,6))
sns.scatterplot(data=housing, x='sqft_living', y='price', alpha=0.3)
sns.regplot(data=housing, x='sqft_living', y='price',
            scatter=False, color='red')
plt.title('Price vs Sqft Living')
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(data=housing, x='sqft_lot', y='price', alpha=0.3)
plt.yscale('log')
plt.xscale('log')
plt.title('Price vs Lot Size (log-log)')
plt.show()
```

```
for col in ['bedrooms', 'bathrooms', 'floors', 'waterfront', 'view', 'grade', 'condition']:
    plt.figure(figsize=(8,5))
    sns.boxplot(data=housing, x=col, y='price')
    plt.title(f'Price by {col}')
    plt.xticks(rotation=45)
    plt.show()
```
